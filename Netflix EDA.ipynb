# 1) Distribution of Netflix TV shows and Movies

```python
# Count of movies and TV shows
data_temp = {'type':['movies','TV_shows'],'count':[6126,2664]}
df_count = pd.DataFrame(data_temp)

# Visualization
sns.set_theme(style = 'whitegrid')

plt.figure(figsize = (8,6))
plt.pie(df_count['count'],
        labels = df_count['type'], 
        explode = (0.03,0.03),
        autopct = '%1.1f%%', 
        startangle = 90, 
        colors= ['red','grey'],
        shadow = True,
        textprops = {'fontweight': 'bold'})
plt.title('Distribution of TV Shows and Movies on Netflix')
plt.axis = ('equal')
plt.show()
```

#### Insights :

* The dataset contains a larger number of movies 69.7% as compared to TV shows which are 30.3% indicating that movies are more
  prevalent on Netflix.
* The Higher Count of movies suggest that Netflix caters to a larger audience interested in on-demand movie content.

# 9) Trend of Content Released Over the Years From 2012 to 2021

```python
filtered_trend = df[(df['release_year']>=2012)& (df['release_year']<=2021)]
release_trend = filtered_trend.groupby('release_year').size()
plt.figure(figsize = (12,6))

plt.plot(release_trend.index, release_trend.values, marker= 'o', color= 'red', label= 'Content Released')
for x,y in zip(release_trend.index, release_trend.values):
    plt.text(x,y,str(y), fontsize=10, ha= 'center',va= 'bottom',color='black')

release_trend.plot(kind= 'line',marker = 'o', color = 'red')
plt.title("Number of Netflix Releases Over the Years")
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.grid()
plt.show()
```

#### Insights
* From the above graph we can see that Netlix's highest content release was in 2018 with 1146 titles, showcasing its peak 
  production strategy.
* The years 2017 and 2019 followed closely with 1030 titles each, while 2020 saw a slight decline to 953 due to pandemic's 
  impact on production.

# 10) Trend of Movies Vs TV Shows Over the years

```python
movies = df[df['type']== 'Movie'].groupby('release_year').size()
tv_shows = df[df['type']== 'TV Show'].groupby('release_year').size()

plt.figure(figsize=(12,6))
plt.plot(movies, label = 'Movies',marker='o')
plt.plot(tv_shows, label = 'TV Show', marker = 's')
plt.title('Netflix Releases: Movies Vs TV Shows')
plt.xlabel('Year')
plt.ylabel('Number of Releaes')
plt.grid()
plt.show()
```

#### Insights:
* The trend of Netflix releases shows that movies consistently outnumber TV shows, indicating a strong focus on delivereing
  a variety of films to cater to diverse audiences.
* While TV shows have grown steadily over the years, movies dominate Netflix's content portfolio.
